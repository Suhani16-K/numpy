NumPy->fundamental package->scientific computing in Python .

which include matrices.
justlike there is datatype called as numpy array.
it is used becoz - 1.⚡ Faster.   2. 📊 Supports Arrays (n-dimensional)   3.➕ Easy Mathematical Operations   4.🧠 Used in ML & AI   5.💾 Memory Efficient


```
import numpy as np
```

```
# np.array
a=np.array([1,2,3])
print(a)
```

```
# 2D and 3D
b=np.array([[1,2,3] , [4,5,6]])
c=np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(b,c)
# print(c)
```

```
# dtype
np.array([1,2,3],dtype=float)
np.array([2,3,4],dtype=bool)
np.array([1,3,6],dtype=complex)
np.array([1,3,6],dtype=bytes)
```

Array Attributes

```
# 1d array -> vector
# 2d array -> matrix
# 3d array -> tensor
# (3,2,2)-> (no.of 2D array,rows,cols)
a1 = np.arange(10,dtype=np.int32)
a2 = np.arange(16,dtype=float).reshape(4,4)
a3 = np.arange(12).reshape(3,2,2)
print(a3)
```

```
# ndim
np.ndim(a1)
np.ndim(a2)
np.ndim(a3)
```

```
# shape
np.shape(a1)
np.shape(a2)
np.shape(a3)
```

```
# size
np.size(a1)
s= np.size(a2)
print(s,a2)
np.size(a3)
```

Stacking

```
a4=np.arange(12).reshape(3,4)
a5=np.arange(12,24).reshape(3,4)
a5
```

Numpy array vs Python lists

```
# speed -> numpy array speed is approx 50 times fast than list
# list
a= [i for i in range(10000000)]
b= [i for i in range(10000000,20000000)]
c=[]
import time
start=time.time()
for i in range(len(a)):
  c.append(a[i]+b[i])
print(time.time()-start)
```

```
# numpy
a=np.array([10000000])
b=np.array([10000000,20000000])
start=time.time()
c=a+b
print(time.time()-start)
```
