                                            NUMPY

NumPy->fundamental package->scientific computing in Python .
It is a Python library that provides a multidimensional array object.
which include matrices.
justlike there is datatype called as numpy array.
it is used becoz - 1.⚡ Faster.   2. 📊 Supports Arrays (n-dimensional)   3.➕ Easy Mathematical Operations   4.🧠 Used in ML & AI   5.💾 Memory Efficient


```
import numpy as np
```

```
# np.array
a=np.array([1,2,3])
print(a)
```

```
# 2D and 3D
b=np.array([[1,2,3] , [4,5,6]])
c=np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(b,c)
# print(c)
```

```
# dtype
np.array([1,2,3],dtype=float)
np.array([2,3,4],dtype=bool)
np.array([1,3,6],dtype=complex)
np.array([1,3,6],dtype=bytes)
```

```
# np.arange
np.arange(1,32,3)
```

```
# with reshape
np.arange(1,13).reshape(2,6)
```

```
# np.ones and np.zeros
# used for initialising the  weight in DL
np.ones((3,4))
np.zeros((3,4))
```

```
# np.random
np.random.random((1,10))
```

```
# np.linspace
np.linspace(-10,10,10)
```

```
# np.identity
np.identity(4)
```

Array Attributes

```
# 1d array -> vector
# 2d array -> matrix
# 3d array -> tensor
# (3,2,2)-> (no.of 2D array,rows,cols)
a1 = np.arange(10,dtype=np.int32)
a2 = np.arange(16,dtype=float).reshape(4,4)
a3 = np.arange(12).reshape(3,2,2)
print(a3)
```

```
# ndim
np.ndim(a1)
np.ndim(a2)
np.ndim(a3)
```

```
# shape
np.shape(a1)
np.shape(a2)
np.shape(a3)
```

```
# size
np.size(a1)
s= np.size(a2)
print(s,a2)
np.size(a3)
```

In NumPy, memory depends on data type (dtype).

Data Type	Size
int16  	   2 bytes   
int32	     4 bytes   
int64	     8 bytes   
float32    4 bytes   
float64    8 bytes   

```
# itemsize
print(a1.itemsize) #when int is 32 and when int is 16 then ->2 byte
print(a2.itemsize) #when it is float
print(a3.itemsize) # when int is 64
```

```
#dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

Changing Datatype

```
a3.astype(np.int32)
```

Array Operations

```
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
print(a1)
print(a2)
```

```
# scalar operations

# arithmetic
a1 ** 2
```

```
# relational
a2 == 15
```

```
# vector operations
# arithmetic
a1 ** a2
```
